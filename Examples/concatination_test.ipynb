```python
import librosa
import librosa.display
import IPython as ip

import audio_plot_util as apu

DATA_DIR = "../Dataset/IRMAS_Training_Data"
```

```python
vocal_path = f"{DATA_DIR}/voi/[voi][pop_roc]2345__2.wav"
acc_guit_path = f"{DATA_DIR}/gac/[gac][cla]0518__1.wav"
```

```python
voc, sr_voc = librosa.load(vocal_path, mono=True,sr=44100)
gac, sr_gac = librosa.load(acc_guit_path, mono=True, sr=44100)
```

```python
ip.display.Audio(voc, rate=sr_voc)
```

```python
ip.display.Audio(gac, rate=sr_gac)
```

```python
apu.plot_signal(voc, sr_voc)
print(f"SR:{sr_voc}")
```

```python
apu.plot_signal(gac, sr_gac)
print(f"SR:{sr_gac}")
```

```python
apu.plot_spectrograms_np(voc, sr_voc)
```

```python
apu.plot_spectrograms_np(voc, sr_voc)
```

```python
# MERGE

result_audio = (voc + gac)/2
result_sr = int((sr_voc + sr_gac)/2)

apu.plot_signal(result_audio, sr=result_sr)

# REPRODUCE
ip.display.Audio(result_audio, rate=result_sr)
```

```python
apu.plot_spectrograms_np(result_audio, result_sr)
```
