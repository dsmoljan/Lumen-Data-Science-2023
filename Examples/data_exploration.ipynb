```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

PATH = "C:/Users/dsmoljan/Desktop/Lumen natjecanje/Dataset/datalists/train.csv"
```

```python
df = pd.read_csv(PATH)
```

```python
features = ["genre", "audio_length", "drums"]

fig, ax = plt.subplots(2, 1)
fig.subplots_adjust(hspace=0.75)

#sns.countplot(x=original_df['fuel'])
for feature, subplot in zip(features, ax.flatten()):
    labels = ["Null percentage", "Non-null percentage"]
    null_percentage = (df[[feature]].isnull().mean() * 100).item()
    non_null_percentage = 100 - null_percentage
    data = [null_percentage, non_null_percentage]
    plt.pie(data, labels = labels, autopct='%.0f%%')
    plt.show()
```
